# Loan Project: Approval Prediction

**Objective** \
Build a model that can accurately predict whether an applicant will be accepted for a loan.

**Purpose** \
I believe the results of this type of project can be used by personal financial advisors to help their clients secure loans. With this model, the advisor can input their clients' information and predict if they are likely to have their application for a loan accepted. ## Missing Values

```python
print(df.isnull().sum())
```

### Handling of Missing Values
- Self_employed: I am assuming that by leaving this unfilled, the applicant is not self-employed. Filling in with "No"
- Gender: This feature is impossible to compute. I could split 50/50, but I think I will just drop the records to avoid biasing the dataset.
- Married: Same as gender; drop null records
- Dependents: Replacing with 0
- LoanAmount: This seems like it could be integral in determining the Loan Status so missing records will be dropped
- Loan_Amount_Term: I will fill with the mode loan term
- Credit_History: Fill with 0

```python
# impute
df['Self_Employed'].fillna("No", inplace=True)
df['Dependents'].fillna("0", inplace=True)
df['Credit_History'].fillna(0, inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)

# drop
df.dropna(inplace=True, subset=['Gender', 'Married', 'LoanAmount'])
```

```python
print("Train:\n", df.isnull().sum())
```

## Feature Engineering

I will be creating the following new features:
- **SingleIncome:** A binary classifier determining if applicant is the only income for their family (dependents >= 1 and coapplicant income = 0)
- **TotalIncome:** Summation of applicant and co-applicant's income (will replace those two columns)

```python
def singInc(df):
    if ((df['CoapplicantIncome'] == 0) and (df['Dependents'] != "0")): return 1
    else: return 0
```

```python
df['SingleIncome'] = df.apply(singInc, axis=1)
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']
# drop applicant and coapplicant income after combining
df.drop(['ApplicantIncome','CoapplicantIncome'], 1, inplace=True)
```

## Shape Change

```python
# here I will use log-transform TotalIncome so that it's shape becomes more Gaussian
from scipy.stats import norm

df['TotalIncome'] = np.log(df['TotalIncome'])  # logarithm of all the values

plt.figure(figsize=(10,6))
sns.distplot(df['TotalIncome'], fit=norm)
```

## Encoding, Dummies, and Scaling

```python
from sklearn.preprocessing import LabelEncoder

# label encode binary
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Married'] = le.fit_transform(df['Married'])
df['Education'] = le.fit_transform(df['Education'])
df['Self_Employed'] = le.fit_transform(df['Self_Employed'])
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])

# get dummies for categorical
df = pd.get_dummies(data=df, columns=['Dependents', 'Property_Area', 'Loan_Amount_Term'])
```

```python
# reset index after dropping in previous steps so I can merge below
df.reset_index(inplace=True, drop=True)

# scale all numerical columns
scaled_features = df.copy()
col_names = ['LoanAmount', 'TotalIncome']
features = scaled_features[col_names]
scaler = StandardScaler()
features = scaler.fit_transform(features.values)
scaled = pd.DataFrame(features, columns=col_names)

# drop unscaled values and merge
df_original = df.copy()
df = df.drop(columns=col_names, axis=1)
df = df.merge(scaled, left_index=True, right_index=True, how='left')
```

## Feature Selection

```python
# drop ID since it's not used for prediction
df.drop('Loan_ID', 1, inplace=True)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.drop('Loan_Status', 1).corr().abs(), cmap='viridis', square=True)
```

```python
def get_collin_feats(df, thresh=0.8):
    # corr matrix in abs val terms
    corr_matrix = df.corr().abs()

    # use only upper triangle of correlations
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

    # select the columns that are above threshold
    collin_features = [column for column in upper.columns if any(upper[column] > thresh)]
    return collin_features
```

```python
get_collin_feats(df, 0.7)
```

No features are strongly correlated so we do not have collinearity here and thus do not need to drop any features.

## Split Data

```python
X = df.drop('Loan_Status', 1)
y = df['Loan_Status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)
```

## Define Plotting Functions

```python
def prerec_curve(y_test, y_score):
    prec_curve, rec_curve, _ = precision_recall_curve(y_test, y_score)
    no_skill = len(y_test[y_test==1]) / len(y_test)
    plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
    plt.plot(rec_curve, prec_curve, marker='.', label='Logistic')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend()
    plt.show()
    # calculate auc
    pr_auc = auc(rec_curve, prec_curve)
    print('AUC=%.3f' % (pr_auc))
    
def roc_curve_do(c_name,classifier,x_test,y_test):
    probs = classifier.predict_proba(x_test)  
    probs = probs[:, 1]  
    fper, tper, thresholds = roc_curve(y_test, probs)
    plt.plot(fper, tper)
    plt.plot([0,1], [0,1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('{} ROC curve'.format(c_name))
    # show the plot
    plt.show()
    # calculate auc
    roc_auc = auc(fper, tper)
    print('AUC=%.3f' % (roc_auc))
    
def metrics(y_test, y_preds):
    Accuracy  =  accuracy_score(y_test, y_preds)
    Precision =  precision_score(y_test, y_preds)
    Recall    =  recall_score(y_test, y_preds)
    F1_score  =  f1_score(y_test, y_preds)
    m = pd.DataFrame(data=[[Accuracy, Precision, Recall, F1_score]], columns=['Accuracy', 'Precision', 'Recall', 'F1_Score'], index=['Model'])
    return m
```

## Modeling

### Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_preds = lr.predict(X_test)
y_score = lr.predict_proba(X_test)[:,1]
```

```python
roc_curve_do('Logistic Regression', lr, X_test, y_test)
prerec_curve(y_test, y_score)
metrics(y_test, y_preds)
```

## Insights

```python
importance = np.std(X, 0).values*lr.coef_
print('Top 5 Most Important Features')
top_feats = pd.DataFrame(importance, columns=X.columns, index=['Importance']).T.sort_values('Importance', ascending=False).head(5)
top_feats
```

Based on these findings, **Credit History** is by far the most important factor in securing a loan.

## Finding a Better Model

### Decision Tree

#### Params Search using GridSearchCV

```python
# define decision tree classifier
dt = DecisionTreeClassifier(random_state=50)

# define parameter grid
params = {
    'min_samples_split' : [2, 3, 4],
    'min_samples_leaf' : [1, 2, 3],
    'min_weight_fraction_leaf': [0.0, 0.25, 0.5, 0.75, 1.0]
}

random_search = GridSearchCV(dt, param_grid=params, scoring='accuracy', n_jobs=-1, cv=5, verbose=3)

random_search.fit(X_train, y_train)
```

```python
dt = random_search.best_estimator_
dt.fit(X_train,y_train)
y_preds = dt.predict(X_test)
y_score = dt.predict_proba(X_test)[:,1]

roc_curve_do('Decision Tree Classifier', dt, X_test, y_test)
prerec_curve(y_test, y_score)
metrics(y_test, y_preds)
```

```python
confusion_matrix(y_test, y_preds)
```

We see here that the decision tree could not perform any better than just predicting every loan to be approved.

### Random Forest Model

#### Params Search using GridSearchCV

```python
# define decision tree classifier
rf = RandomForestClassifier(random_state=50, n_jobs=-1)

# define parameter grid
params = {
    'n_estimators': [50, 100, 250],
    'min_samples_split' : [2, 3, 4],
    'min_samples_leaf' : [1, 2, 3],
    'min_weight_fraction_leaf': [0.0, 0.25, 0.5, 0.75]
}

random_search = GridSearchCV(rf, param_grid=params, scoring='accuracy', n_jobs=-1, cv=5, verbose=3)

random_search.fit(X_train, y_train)
```

```python
rf = random_search.best_estimator_
rf.fit(X_train,y_train)
y_preds = rf.predict(X_test)
y_score = rf.predict_proba(X_test)[:,1]

roc_curve_do('Random Forest Classifier', rf, X_test, y_test)
prerec_curve(y_test, y_score)
metrics(y_test, y_preds)
```

### KNeighbors Classifier

```python
kn = KNeighborsClassifier(n_jobs=-1)

# define parameter grid
params = {
    'n_neighbors': [5, 10, 15],
    'leaf_size' : [5, 10, 15],
}

random_search = GridSearchCV(kn, param_grid=params, scoring='accuracy', n_jobs=-1, cv=5, verbose=3)

random_search.fit(X_train, y_train)
random_search.best_estimator_
```

```python
kn = random_search.best_estimator_
kn.fit(X_train,y_train)
y_preds = kn.predict(X_test)
y_score = kn.predict_proba(X_test)[:,1]

roc_curve_do('KNearest Neighbors Classifier', kn, X_test, y_test)
prerec_curve(y_test, y_score)
metrics(y_test, y_preds)
```

## Best Model

```python
models = {
        'LogisticRegression': lr,
        'DecisionTreeClassifier': dt,
        'RandomForestClassifier': rf,
        'KNeighborsClassifier': kn
          }
```

```python
def predict(models, X_test, y_test):
    final_results = pd.DataFrame(columns=['Model', 'Accuracy','Precision','Recall','F1_score'])
    for name, model in models.items():
        y_preds = model.predict(X_test)
        acc =  accuracy_score(y_test, y_preds)
        pre =  precision_score(y_test, y_preds)
        rec =  recall_score(y_test, y_preds)
        f1 =  f1_score(y_test, y_preds)
        final_results = final_results.append({'Model':name, 'Accuracy': acc, 'Precision':pre, 'Recall':rec, 'F1_score':f1}, ignore_index=True)
    return final_results.set_index('Model')
```

```python
predict(models, X_test, y_test)
```

If we choose our model based on accuracy, no model beats our baseline of **Logistic Regression** (accuracy = 75.7%).
